```python
import math
import ipywidgets as widgets
import matplotlib.pyplot as plt
import pandas as pd

A_lists=list()
Bc_lists=list()
N_lists=list()
A=list()
N=list()

print("input \'000\' to stop adding  ")
while (True):
    inp=input("Enter A: ")
    if (inp=="000"):
        break
    else:
        A.append(float(inp))

        
inp2=input("Enter N: ")
for i in range(len(A)):
    N.append(int(inp2))   

def onck(n):
    if(n<0):
        n=0
        print("There's no factorial for a negative number ")
    else:
        global N
        global A
        global A_lists
        global Bc_lists
        global N_lists
        R_list=list()
        t=[]
        N=t
        for i in range(0,len(A)):
              N.append(n)
        R_list=[ ((a**n)/math.factorial(n))/bas(a,n) for a,n in zip(A,N)]
        if R_list not in Bc_lists:
                A_lists.append(A)
                Bc_lists.append(R_list)
                N_lists.append(N)
            
        
        dic={'Bc':R_list,'A':A,'N':N}
        df=pd.DataFrame(dic)
        plotting(A_lists,Bc_lists,n,N_lists)
        plt.grid(True)
        plt.xlabel("A")
        plt.ylabel('Bc')
        plt.axis([0.1,100,0.001,0.1])
        plt.loglog()
        plt.grid(True)
        plt.legend()
        plt.show()
        print(df.to_string(index=False))
                
def bas(a,n):
    b=0
    for i in range(n+1):
        b+=a**i/math.factorial(i)
    return b
def plotting(A_ls,Bc_ls,n,N):
    for i in range(len(Bc_ls)):
        plt.plot(A_ls[i],Bc_ls[i],label='N= '+str(N[i][0]))
        


tex=widgets.IntText(
    value=int(inp2),
    description='N:',
    disabled=False,
    continuous_update=True
)

widgets.interact(onck,n=tex)
```
